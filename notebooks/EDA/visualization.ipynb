# Data Visualization - Movie Recommendation System

This notebook provides visual insights into the movie dataset, including rating distributions, genre popularity, and feature correlations.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast
import warnings
warnings.filterwarnings('ignore')

sns.set(style="whitegrid")
```

## 2. Load Data

```python
df = pd.read_csv('../../data/final/sample.csv')
# Convert string representation of lists to actual lists
df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

## 3. Visualization: Vote Average Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['vote_average'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Vote Average')
plt.xlabel('Vote Average')
plt.ylabel('Frequency')
plt.show()
```

## 4. Visualization: Top 10 Genres

```python
all_genres = [genre for sublist in df['genres'].dropna() for genre in sublist]
genre_counts = pd.Series(all_genres).value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='viridis')
plt.title('Top 10 Movie Genres')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```

## 5. Visualization: Correlation Heatmap

```python
plt.figure(figsize=(10, 8))
numerical_df = df[['vote_average', 'vote_count', 'revenue', 'runtime', 'budget']]
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

## 6. Visualization: Budget vs Revenue

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='budget', y='revenue', alpha=0.5, color='coral')
plt.title('Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```
